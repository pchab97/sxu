```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
```

```python
cars = pd.read_csv('samochody1tys.csv')
print(cars)
```

```python
x = np.linspace(-3.14, 3.14, 100)
tg = np.sin(x)

pl.xlabel('Oś X')
pl.ylabel('Oś Y')
pl.title("Pierwszy wykres")

pl.plot(x, x, label='funkcja liniowa')
pl.plot(x, x**2, label='funkcja kwadratowa')
pl.plot(x, x**3, label='funkcja trzeciego rzędu')
pl.plot(x, tg, label='funkja trygonometryczna - sinus')

pl.legend()
pl.show()
```

```python
carss = cars.iloc[:, 0:2]
cars_group = carss.groupby('marka').count()
cars_copy = cars_group.copy()
cars_copy2 = cars_copy[cars_copy['id'] >= 50.0] 
print(cars_copy2)
bars = cars_copy2.iloc[:, 0:2].count()["id"]

N = np.arange(bars)
carsAmount = cars_copy2["id"]
width = 0.8
p = pl.bar(N, carsAmount, width, 0)
pl.ylabel('Ilość')
pl.xticks(N,('Audi','BMW','Ford','Mercedes-Benz','Opel','Renault','Volkswagen'))
pl.title('Liczebność marek samochodów')

pl.show()
```

```python
cars2 = cars.iloc[:, [0,3]]
cars2_group = cars2.groupby('rok_produkcji').count()
cars3_group = cars2_group[cars2_group['id'] >= 50.0]
print(cars3_group)
labels = '2005','2006','2007','2008','2009','2014','2016','2017'
sizes = cars3_group["id"]
explode = (0.1, 0.2, 0.2, 0.1, 0.2, 0.2, 0.2, 0.1,)

fig1, ax1 = pl.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',shadow=True, startangle=0)
ax1.axis('equal')

pl.show()
```

```python
cars5 = cars.iloc[:, [7]].to_numpy()

pl.hist(cars5, bins= 10, alpha=1, rwidth=1)

pl.xlabel('Cena')
pl.ylabel('Ilość')
pl.title('Ceny samochodów')
```
